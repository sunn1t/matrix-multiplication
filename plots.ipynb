```python
# Чтобы определить размер матрицы, при котором нужно прерывать рекурсию
# в алгоритме Штрассена, будем запускать оба алгоритма на квадратных матрицах
# с размером стороны 2, 4, 8 и т.д. 

# В первом измерении размер матрицы 2, при перемножении матриц размера 1 
# используется умножение матриц по определению

# Далее, если оказывается, что матрицы размера n на n перемножаются по
# определению быстрее, чем в алгоритме Штрассена, то для матриц размера
# 2*n на 2*n рекурсия прерывается после первого шага, то есть матрицы
# n на n уже перемножаются по определению
```

```python
import matplotlib.pyplot as plt

matrix_size = [2, 4, 8, 16, 32, 64]
time_mul = [2.78e-06, 4.28e-06, 1.344e-05, 7.291e-05, 5.504e-04, 3.401e-03]
strassen_time = [5.29e-05, 6.89e-05, 1.060e-04, 2.181e-04, 8.612e-04, 3.698e-03]

plt.plot(matrix_size, time_mul)
plt.plot(matrix_size, strassen_time)
plt.ylabel('Время, c')
plt.xlabel('Размер матрицы')
plt.legend ( ("По определению", "Алгоритм Штрассена") )
plt.show()
```

```python
# Если размер матрицы не превышает 64 на 64, алгоритм Штрассена проигрывает по эффективности
# Для этих и следующих двух значений время работы измерялось 100 раз, чтобы определить, когда
# алгоритм Штрассена станет быстрее, чем умножение по определению

# Далее время работы измерялось по 10 раз

# При размере 128 на 128, если прерывать рекурсию при умножении матриц 64 на 64, алгоритм Штрассена уже
# становится эффективнее

# Среднее время работы на матрицах 128 на 128:
# 29.66 мс для обычного умножения
# 28.14 мс для алгоритма Штрассена

# Значит выгодно прерывать рекурсию, когда размер матриц стал равен 64
```

```python
import matplotlib.pyplot as plt

matrix_size = [64, 128, 256, 512, 1024, 2048, 4096]
time_mul = [3.401e-03, 0.02966, 0.2398, 2.030, 19.416, 210.615, 1875.16]
strassen_time = [3.698e-03, 0.02814, 0.1917, 1.352, 9.908, 69.352, 486.018]

plt.plot(matrix_size, time_mul)
plt.plot(matrix_size, strassen_time)
plt.ylabel('Время, c')
plt.xlabel('Размер матрицы')
plt.legend ( ("По определению", "Алгоритм Штрассена") )

plt.xlim (100, 600)
plt.ylim (0, 3)

plt.show()

matrix_size = [64, 128, 256, 512, 1024, 2048, 4096]
time_mul = [3.401e-03, 0.02966, 0.2398, 2.030, 19.416, 210.615, 1875.16]
strassen_time = [3.698e-03, 0.02814, 0.1917, 1.352, 9.908, 69.352, 486.018]

plt.plot(matrix_size, time_mul)
plt.plot(matrix_size, strassen_time)
plt.ylabel('Время, c')
plt.xlabel('Размер матрицы')
plt.legend ( ("По определению", "Алгоритм Штрассена") )

plt.xlim (500, 4100)
plt.ylim (1, 2000)

plt.show()
```
